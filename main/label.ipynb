```python
filename = "harvey_twitter_dataset/02_archive_only/HurricaneHarvey.json"
i = 0
with open('output.csv','w',encoding="UTF-8") as g:
    g.write('id^text')
    g.write('\n')
    with open(filename, 'r',encoding="UTF-8") as f:
        for line in f:
            a = time.time()
            data = json.loads(line)
            g.write('%s^%s' % (str(data['id']),str(data['text'])))
            g.write('\n')
            i += 1
            if i % 10000 == 0:
                b = time.time()
                print('%i/%i finished (%.8fs per tweet)' % (i,7041866,(b-a)/10000))
```

```python
filename = "harvey_twitter_dataset/02_archive_only/HurricaneHarvey.json"
i = 0
with open('output.csv', 'w', newline='') as csvfile:
    with open(filename, 'r',encoding="UTF-8") as f:
        writer = csv.writer(csvfile, delimiter='=')
        for line in f:
            a = time.time()
            data = json.loads(line)
            writer.writerow([str(data['id']),str(data['text']).strip()])
            i += 1
            if i % 10000 == 0:
                b = time.time()
                print('%i/%i finished (%.8fs per tweet)' % (i,7041866,(b-a)/10000))
```
